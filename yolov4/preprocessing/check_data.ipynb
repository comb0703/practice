```python
import pandas as pd
import numpy as np
import json
import os

path = './json/'
os.listdir(path)

```

```python
data_list = []

for i in os.listdir(path) :
  target_json = './json/' + i
  with open(target_json, "r", encoding="utf-8") as f:
    data = json.load(f)
    data_list.append(data)
```

```python
# select which data
data = data_list[0]
```

```python
# categories
categories = pd.DataFrame(data['categories'])
print(categories)
categories.isnull().sum()
```

```python
# images
images = pd.DataFrame(data['images'])
print(images)
images.isnull().sum()
```

```python
# annotations
annotations = pd.DataFrame(data['annotations'])
print(annotations)
annotations.isnull().sum()
```

```python
from collections import Counter

subclass = Counter(annotations['category_id'])
subclass = list(subclass.items())
subclass.sort(key=lambda subclass : subclass[0])
```

```python
subclass = pd.DataFrame(subclass,columns=["sub_idx","num"])
subclass = pd.concat([categories.name,subclass],axis=1)
subclass = subclass.fillna(0)
subclass[['sub_idx', 'num']] = subclass[['sub_idx', 'num']].astype(int)
subclass.to_csv('current.csv')

```

```python
# 기본정보
print("해당 슈퍼클래스의 총 이미지의 개수는 : ",images.shape[0])
print("해당 슈퍼클래스의 서브클래스 개수는 : ",categories.shape[0])
print("해당 슈퍼클래스의 어노테이션 개수는 : ",annotations.shape[0])
num_of_img = (annotations.shape[0]/images.shape[0])
num_of_img = round(num_of_img,2)
print("해당 슈퍼클래스의 이미지 당 평균 객체 개수는 : ", num_of_img)
total = 0

for i in images['height'] :
    if i != 2100 :
        total += 1
print("height != 2100 의 개수는 : ", total)

for i in images['width'] :
    if i != 2800 :
        total += 1
print("width != 2800 의 개수는 : ", total)
```

```python
# 라벨링 bbox 오기

total = 0
for i in images['id'] :
    query = annotations['image_id'] == i
    for j in annotations[query]['bbox'] :
        if j == [0.0,0.0,0.0,0.0] :
            total += 1
            query = images['id'] == i
            print(images[query]['file_name'])
print('총', total,"개")
```

```python
# 라벨링 area 오기

total = 0
for i in images['id'] :
    query = annotations['image_id'] == i
    for j in annotations[query]['area'] :
        if j == 0.0 :
            total += 1
            query = images['id'] == i
            print(images[query]['file_name'])
print('총', total,"개")
```

```python
# 라벨링 누락

total = 0
for i in images['id'] :
    query = annotations['image_id'] == i
    if len(annotations[query]) == 0 :
        total += 1
        query = images['id'] == i
        print(images[query]['file_name'])

print('총', total,"개")
```

```python
# area 범위 설정
total = []
for i in images['id'] :
    query = annotations['image_id'] == i
    for area in annotations[query]['area'] :
        if area <= 40000.0 :
            query = images['id'] == i
            name = images[query]['file_name']
            name = list(name)
            total.extend(name)
print('총', len(total),"개")
total
```

```python
# bbox 범위 설정
total = []
for i in images['id'] :
    query = annotations['image_id'] == i
    for bbox in annotations[query]['bbox'] :
        if bbox[2] <= 200.0 and bbox[3] <= 200.0:
            query = images['id'] == i
            name = images[query]['file_name']
            name = list(name)
            total.extend(name)
print('총', len(total),"개")
total
```
