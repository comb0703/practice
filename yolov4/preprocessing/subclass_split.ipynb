```python
import pandas as pd

data = pd.read_csv("./nia_total.csv",index_col=[0])
```

```python
# 23 classes, 0,0 보다 큰
a = [18, 50, 53, 59, 60, 61, 62, 63, 70, 72, 84, 90, 95, 97, 104, 106, 107, 109, 110, 111, 112, 113, 127] 
# 18 classes, 0.01 보다 큰
b = [50, 59, 60, 61, 62, 63, 70, 72, 90, 95, 97, 104, 106, 107, 110, 112, 113, 127]
# 9 classes, 0.1 보다 큰
c = [50, 59, 61, 95, 107, 110, 112, 113, 127]

# select training class
query = data['train_cls_idx'].isin(b)
data = data[query]
```

```python
data = data.sort_values(by='train_cls_idx')
data
```

```python
from collections import Counter

class_num = Counter([c for c in data['train_cls_idx']])
print('총 클래수 개수 :',len(class_num))
```

```python
total = []
index_list = []
for a,b in class_num.items() :
  total.append(b)
  index_list.append(a)

print('총 어노테이션 :', sum(total))
```

```python
images_num = Counter([c for c in data['train_img_idx']])
images_num 
print('이미지 개수 :',len(images_num))
```

```python
# split t,v,t in each class (7:1:2) annotation 기준
split = []
num = max(class_num.keys())+1

for i in range(num) :
  if class_num[i] > 10 :
    train_num = (class_num[i] // 10) * 7
    val_num = (class_num[i] // 10) * 1
    test_num = class_num[i] -train_num - val_num
  else :
    continue
  split.append([train_num,val_num,test_num])

split
```

```python
total_split = []
for s in split :
  for a in range(s[0]) :
    total_split.append('train')
  for b in range(s[1]) :
    total_split.append('val')
  for c in range(s[2]) :
    total_split.append('test')
# annotation 개수와 동일해야 함
len(total_split)
```

```python
# annotation 기준 split을 image 기준 split으로 변환
cur = -1
idx = -1
final = []
for i in data['train_img_idx']:
  if cur != i :
    idx += 1
    final.append(total_split[idx])
    cur = i
  else : 
    final.append(total_split[idx])
len(final)
```

```python
data['split'] = final
```

```python
# 이미지 파일명 json read
name = pd.read_csv('total_imgs.csv',index_col=[0])
name
```

```python
# 이미지 파일명 가져와서 list 만듬
names= []
# 새로운 인덱스와 맵핑
map = pd.read_csv('./NIA_1103_CATE.csv',encoding='utf-8')
new_class = []
for idx, val in data.iterrows():
  query = (name['json_idx'] == val['json_idx']) & (name['id'] == val['image_id'])
  names.extend(list(name[query]['file_name']))
  
  query = map['init_idx'] == val['train_cls_idx']
  new_class.extend(list(map[query]['name']))

```

```python
data['file_name'] = names
data['name'] = new_class
```

```python
c = Counter([c for c in data['name']])
d = {}
for index,c in enumerate(c.keys()) :
    d[c] = index
ori = []
for idx, val in data.iterrows():
    ori.append(d[val['name']])
```

```python
data['index'] = ori
```

```python
data = data[['index','name','train_cls_idx','train_img_idx','x','y','w','h','split','file_name']]
data.reset_index(inplace=True, drop=True) 
```
